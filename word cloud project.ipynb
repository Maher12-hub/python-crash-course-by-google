```python
import wordcloud
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import display
```

```python
file_contents=['Maher','Rihon','Saifa','Mehanaz']
def calculate_frequency(file_contents):
    words=file_contents
    word_count={}
    for word in words:
        if word not in word_count:
            word_count[word]=0
        word_count[word]+=1
    cloud=wordcloud.WordCloud()
    cloud.generate_from_frequencies(word_count)
    return cloud.to_array()
my_image=calculate_frequency(file_contents)
plt.imshow(my_image,interpolation='nearest')
plt.axis('off')
plt.show()
```

```python
import wordcloud
from matplotlib import pyplot as plt
from IPython.display import display


with open('wordcloud.txt','r') as f:
    file_contents=f.read()
def calculate_frequency(file_contents):
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just"]
    words=file_contents.split()
    word_count={}
    for word in words:
        if word in uninteresting_words:
            continue
        else:
            if word not in word_count:
                word_count[word]=0
            word_count[word]+=1
    cloud=wordcloud.WordCloud()
    cloud.generate_from_frequencies(word_count)
    return cloud.to_array()
my_image=calculate_frequency(file_contents)
plt.imshow(my_image,interpolation='nearest')
plt.axis('off')
plt.show()
```
